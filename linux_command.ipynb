# cd
`cd` (Change Directory)는 현재 작업 중인 디렉토리를 변경하는 명령어입니다.

**주요 사용법:**
*   `cd [디렉토리 경로]`: 지정된 디렉토리로 이동합니다.
*   `cd ..`: 현재 디렉토리의 상위 디렉토리로 이동합니다.
*   `cd ~` 또는 `cd`: 현재 사용자의 홈 디렉토리로 이동합니다.
*   `cd -`: 이전에 작업했던 디렉토리로 이동합니다.

```python
!cd ..
```

#파일 허가권
총 9칸, 3칸씩 끊어서 읽음

*   r: read, w: write, x: execute
*   첫 번째 셋은 소유자(User)의 파일접근 권한
*   두 번째 셋은 그룹(Group)의 파일접근 권한
*   세 번째 셋은 그 외의 사용자(Other)의 파일접근 권한





#chmod
파일 허가권을 변경함

*   chmod 777 sample.txt
*   chmod 허가권 허가권_변경할_파일명




#chown

*   단, root계정에게 소유권 넘기기는 불가
