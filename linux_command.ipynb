## ps -ef | grep [프로세스명]

이 명령어는 현재 시스템에서 실행 중인 특정 프로세스를 찾는 데 사용됩니다.

*   **`ps -ef`**: 현재 실행 중인 모든 프로세스에 대한 자세한 정보를 표시합니다.
    *   `-e`: 모든 프로세스를 선택합니다.
    *   `-f`: 전체 형식(full format)으로 정보를 표시합니다. (UID, PID, PPID, C, STIME, TTY, TIME, CMD 등)

*   **`|` (파이프)**: `ps -ef` 명령어의 출력을 `grep` 명령어의 입력으로 전달합니다.

*   **`grep [프로세스명]`**: 이전 명령어로부터 전달받은 텍스트에서 특정 문자열(`[프로세스명]`)을 포함하는 라인을 필터링하여 출력합니다.

**예시:**
```bash
ps -ef | grep sshd
```
이 명령어는 `sshd` (SSH 데몬)라는 문자열을 포함하는 프로세스를 찾아 그 정보를 출력합니다.

## 파이썬 소켓 채팅 프로그램 (서버 & 클라이언트)

이 예제는 파이썬의 `socket` 모듈을 사용하여 간단한 서버와 클라이언트 간의 통신을 보여줍니다. 서버는 클라이언트의 연결을 기다리고, 클라이언트가 메시지를 보내면 응답합니다.

### 1. 서버 코드 (Server Code)

아래 코드를 실행하여 서버를 시작합니다. 서버는 클라이언트의 연결을 기다리며, 연결이 되면 메시지를 주고받습니다. 이 셀을 먼저 실행해야 합니다. **Colab 환경에서는 서버가 연결을 기다리는 동안 셀이 계속 실행 중인 상태로 유지됩니다. 클라이언트가 연결되기 전까지 다음 셀이 실행되지 않으니, 이 셀을 실행한 후 다른 Colab 탭이나 로컬 환경에서 클라이언트 코드를 실행해주세요.**

````python
import socket

HOST = '127.0.0.1'  # 표준 루프백 인터페이스 주소 (localhost)
PORT = 65432        # 서버가 수신 대기할 포트 (비특권 포트는 1023보다 큼)

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.bind((HOST, PORT))
    s.listen()
    print(f"서버가 {HOST}:{PORT} 에서 수신 대기 중입니다.")
    conn, addr = s.accept()
    with conn:
        print(f"클라이언트가 {addr} 에서 연결되었습니다.")
        data = conn.recv(1024) # 클라이언트로부터 1024 바이트까지 데이터 수신
        if data:
            print(f"클라이언트로부터 받은 메시지: {data.decode('utf-8')}")
            # 한글 문자열을 UTF-8로 인코딩하여 전송
            conn.sendall("안녕하세요, 클라이언트! 메시지를 잘 받았습니다.".encode('utf-8'))
        else:
            print("클라이언트가 데이터를 보내지 않고 연결을 끊었습니다.")
print("서버 종료.")
````

### 2. 클라이언트 코드 (Client Code)

서버 코드를 실행한 후, **다른 Colab 탭이나 별도의 터미널에서 아래 클라이언트 코드를 실행**하여 서버에 연결하고 메시지를 주고받습니다. 클라이언트가 실행되면 서버에 메시지를 보내고 서버의 응답을 받습니다.

````python
import socket

HOST = '127.0.0.1'  # 서버의 호스트 이름 또는 IP 주소
PORT = 65432        # 서버가 사용하는 포트

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.connect((HOST, PORT))
    print(f"서버 {HOST}:{PORT} 에 연결되었습니다.")
    # 한글 문자열을 UTF-8로 인코딩하여 전송
    s.sendall("안녕하세요, 서버! 클라이언트입니다.".encode('utf-8'))
    data = s.recv(1024) # 서버로부터 1024 바이트까지 데이터 수신
    print(f"서버로부터 받은 메시지: {data.decode('utf-8')}")
print("클라이언트 종료.")
````

### Colab 환경에서 소켓 통신 문제 설명

`127.0.0.1` (localhost)은 현재 실행 중인 컴퓨터 자체를 의미합니다. Colab 환경에서 각 탭이나 세션은 별도의 격리된 가상 머신(VM) 또는 컨테이너에서 실행될 수 있습니다. 따라서 한 Colab 탭에서 실행되는 서버가 다른 Colab 탭에서 실행되는 클라이언트와 `127.0.0.1`을 통해 직접 통신할 수 없습니다. 각각의 `127.0.0.1`은 서로 다른 가상 머신을 가리키기 때문입니다.

이를 해결하고 Colab 내에서 소켓 통신을 시연하려면, 서버와 클라이언트 코드를 동일한 Python 프로세스 내에서 동시에 실행해야 합니다. Python의 `threading` 모듈을 사용하면 이를 구현할 수 있습니다.

### 3. 단일 Colab 셀에서 스레딩을 사용한 서버-클라이언트 예제

아래 코드는 `threading` 모듈을 사용하여 하나의 Colab 셀 내에서 서버와 클라이언트를 동시에 실행하는 예제입니다. 서버는 별도의 스레드에서 시작되고, 그 즉시 클라이언트가 서버에 연결하여 메시지를 주고받습니다. 이 예제에서는 `127.0.0.1`을 사용하여 동일한 런타임 내에서의 통신이 가능합니다.

````python
import socket
import threading
import time

HOST = '127.0.0.1'
PORT = 65432

def server_program():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind((HOST, PORT))
        s.listen()
        print(f"[서버] {HOST}:{PORT} 에서 수신 대기 중입니다.")
        conn, addr = s.accept() # 클라이언트 연결 대기
        with conn:
            print(f"[서버] 클라이언트가 {addr} 에서 연결되었습니다.")
            data = conn.recv(1024)
            if data:
                print(f"[서버] 클라이언트로부터 받은 메시지: {data.decode('utf-8')}")
                response = "안녕하세요, 클라이언트! 메시지를 잘 받았습니다."
                conn.sendall(response.encode('utf-8'))
            else:
                print("[서버] 클라이언트가 데이터를 보내지 않고 연결을 끊었습니다.")
    print("[서버] 종료.")

def client_program():
    # 서버가 시작될 시간을 주기 위해 잠시 대기
    time.sleep(1)
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((HOST, PORT))
        print(f"[클라이언트] 서버 {HOST}:{PORT} 에 연결되었습니다.")
        message = "안녕하세요, 서버! 클라이언트입니다."
        s.sendall(message.encode('utf-8'))
        data = s.recv(1024)
        print(f"[클라이언트] 서버로부터 받은 메시지: {data.decode('utf-8')}")
    print("[클라이언트] 종료.")


if __name__ == '__main__':
    print("### 스레드를 이용한 소켓 통신 시작 ###")
    # 서버 스레드 시작
    server_thread = threading.Thread(target=server_program)
    server_thread.start()

    # 클라이언트 스레드 시작
    client_thread = threading.Thread(target=client_program)
    client_thread.start()

    # 두 스레드가 모두 종료될 때까지 대기
    server_thread.join()
    client_thread.join()
    print("### 스레드를 이용한 소켓 통신 완료 ###")
````

## /etc/fstab

`/etc/fstab` 파일은 Linux 시스템에서 파일 시스템을 마운트하는 정보를 포함하는 설정 파일입니다. 시스템이 부팅될 때 자동으로 마운트될 파일 시스템, 장치, 마운트 지점, 파일 시스템 유형, 마운트 옵션 등이 정의됩니다.

**주요 필드:**

1.  **File System (장치명 또는 UUID):** 마운트할 장치의 경로 또는 UUID를 지정합니다. (예: `/dev/sdb1` 또는 `UUID=xxxxxxxx-xxxx-xxxx-xxxx-xxxxxxxxxxxx`)
2.  **Mount Point (마운트 지점):** 장치를 마운트할 디렉토리 경로입니다. (예: `/mnt/data`)
3.  **Type (파일 시스템 종류):** 파일 시스템의 유형을 지정합니다. (예: `ext4`, `xfs`, `ntfs`, `swap`)
4.  **Options (마운트 옵션):** 파일 시스템 마운트 시 사용할 옵션들을 쉼표로 구분하여 지정합니다. (예: `defaults`, `noatime`, `ro`, `rw`, `noauto`)
    *   `defaults`: `rw, suid, dev, exec, auto, nouser, async` 옵션을 포함합니다.
    *   `noatime`: 파일 접근 시간을 업데이트하지 않아 디스크 I/O를 줄입니다.
    *   `ro`: 읽기 전용으로 마운트합니다.
    *   `rw`: 읽기/쓰기 가능으로 마운트합니다.
    *   `noauto`: 부팅 시 자동으로 마운트하지 않습니다.
5.  **Dump (덤프 여부):** `dump` 명령어로 백업할지 여부를 지정합니다. `0`이면 백업하지 않고, `1`이면 백업합니다.
6.  **Pass (fsck 순서):** `fsck` 명령어로 파일 시스템 무결성 검사를 수행할 순서를 지정합니다. `0`이면 검사하지 않고, `1`은 루트 파일 시스템, `2`는 그 외의 파일 시스템입니다.

## 간단한 쉘 스크립트 계산기

아래 코드는 두 개의 숫자와 하나의 연산자(`+`, `-`, `*`, `/`)를 입력받아 계산을 수행하는 간단한 쉘 스크립트 계산기입니다.

### 사용 방법:
1.  아래 셀을 실행하여 스크립트를 정의합니다.
2.  `calculate 10 + 5` 와 같이 `calculate` 명령 뒤에 숫자, 연산자, 숫자를 차례로 입력하여 사용합니다.

````python
%%bash

# calculate 함수 정의
calculate() {
  if [ "$#" -ne 3 ]; then
    echo "사용법: calculate <숫자1> <연산자> <숫자2>"
    echo "예시: calculate 10 + 5"
    return 1
  fi

  num1=$1
  operator=$2
  num2=$3

  # 입력값이 숫자인지 확인
  if ! [[ "$num1" =~ ^[0-9]+$ ]] || ! [[ "$num2" =~ ^[0-9]+$ ]]; then
    echo "오류: 숫자만 입력해주세요."
    return 1
  fi

  result=0
  case $operator in
    "+")
      result=$((num1 + num2))
      ;;
    "-")
      result=$((num1 - num2))
      ;;
    "*")
      result=$((num1 * num2))
      ;;
    "/")
      if [ "$num2" -eq 0 ]; then
        echo "오류: 0으로 나눌 수 없습니다."
        return 1
      fi
      result=$((num1 / num2))
      ;;
    *)
      echo "오류: 지원하지 않는 연산자입니다. (+, -, *, / 만 사용 가능)"
      return 1
      ;;
  esac

  echo "결과: $result"
}

# 함수를 전역으로 내보내어 다른 쉘 명령에서 사용 가능하게 함
export -f calculate

echo "계산기 함수 'calculate'가 정의되었습니다."
````

## 간단한 텔넷 서버 구축 (xinetd, telnetd 활용)

텔넷(Telnet)은 원격 호스트에 접속할 때 사용하는 오래된 프로토콜입니다. 사용자 인증 정보와 통신 내용이 암호화되지 않고 평문으로 전송되기 때문에 보안에 매우 취약하며, 실제 운영 환경에서는 SSH와 같은 안전한 프로토콜 사용이 강력히 권장됩니다. 이 설명은 학습 및 테스트 목적으로만 활용하시기 바랍니다.

`xinetd`는 인터넷 서비스 데몬으로, 요청이 있을 때만 특정 서비스를 시작하여 시스템 리소스를 효율적으로 관리하는 역할을 합니다. `telnetd`는 텔넷 서비스를 제공하는 데몬입니다.

### 1. `xinetd` 및 `telnetd` 설치

먼저 필요한 패키지들을 설치해야 합니다. Colab 환경에서는 `apt`를 사용하여 설치합니다.

````python
%%bash
sudo apt update
sudo apt install -y xinetd telnetd
````

### 2. `telnet` 서비스 설정 (xinetd)

`xinetd`가 `telnet` 서비스를 관리하도록 설정 파일을 생성하거나 수정해야 합니다. `/etc/xinetd.d/telnet` 파일을 생성하고 다음 내용을 추가합니다.

````python
%%bash
sudo tee /etc/xinetd.d/telnet > /dev/null << EOF
service telnet
{
    disable         = no
    flags           = REUSE
    socket_type     = stream
    wait            = no
    user            = root
    server          = /usr/sbin/in.telnetd
    log_on_failure  += USERID
}
EOF

echo "/etc/xinetd.d/telnet 파일이 생성되었습니다."
cat /etc/xinetd.d/telnet
````

### 3. `xinetd` 서비스 재시작

설정 변경 사항을 적용하기 위해 `xinetd` 서비스를 재시작합니다.

````python
%%bash
sudo systemctl restart xinetd
sudo systemctl status xinetd
````

### 4. 텔넷 연결 테스트

이제 텔넷 서버가 제대로 작동하는지 테스트할 수 있습니다. **주의: Colab 환경은 격리되어 있으므로, `localhost` (127.0.0.1)로의 텔넷 접속은 서버 코드와 동일한 원리로, 동일한 런타임 내에서만 가능합니다. 외부에서 Colab VM으로 직접 텔넷 접속은 일반적으로 불가능합니다.**

만약 로컬 환경에서 테스트하고 있다면, 다른 터미널에서 `telnet 127.0.0.1` 또는 `telnet [Colab VM의 외부 IP]` (Colab에서 VM 외부 IP를 얻는 것은 복잡하며, 일반적으로 직접 접근이 차단됨)를 실행해볼 수 있습니다.

Colab 내에서 간단히 텔넷 서비스를 확인하는 방법은 없지만, `netstat` 명령을 통해 텔넷 포트(23번)가 열려 있는지 확인할 수 있습니다.

````python
%%bash
sudo netstat -tulnp | grep 23
````

만약 위 `netstat` 명령의 결과에 `tcp`와 `:23`이 포함된 라인이 보인다면, 텔넷 서비스가 포트 23번에서 수신 대기 중임을 의미합니다.

## OpenSSH 서버 구축

OpenSSH는 원격으로 안전하게 서버에 접속할 수 있도록 해주는 프로토콜인 SSH(Secure Shell)의 오픈 소스 구현체입니다. Telnet과 달리 모든 통신 내용이 암호화되어 전송되므로 보안에 매우 강력하며, 현대 리눅스 시스템에서 원격 접속의 표준으로 사용됩니다.

```bash
sudo apt update
sudo apt install -y openssh-server
```

설치 후 OpenSSH 서비스가 자동으로 시작되는 경우가 많지만, `systemctl`을 사용하여 상태를 확인하고 필요하면 시작할 수 있습니다.

sudo systemctl start ssh
```

### 3. 방화벽 설정 (선택 사항, Colab에서는 불필요)

```

### 4. SSH 설정 파일 (`/etc/ssh/sshd_config`)

OpenSSH 서버의 설정은 `/etc/ssh/sshd_config` 파일에서 관리합니다. 여기서 포트 번호, 인증 방식, 허용 사용자 등을 설정할 수 있습니다. 변경 후에는 SSH 서비스를 재시작해야 적용됩니다.

주요 설정:
*   `Port 22`: SSH 서버가 수신 대기할 포트 번호 (기본값: 22)
*   `PermitRootLogin prohibit-password`: root 계정의 직접 로그인을 비활성화하고 키 기반 인증만 허용
*   `PasswordAuthentication no`: 비밀번호 인증 비활성화 (보안 강화를 위해 권장)
*   `PubkeyAuthentication yes`: 공개키 인증 활성화

```bash
sudo systemctl restart ssh
```

## `ngrok`을 이용한 웹 서버 외부 노출

`ngrok`은 로컬 환경에서 실행 중인 웹 서버를 인터넷에 안전하게 노출시키는 터널링 도구입니다. 이를 통해 Colab VM의 Apache 웹 서버에 외부 웹 브라우저에서 직접 접근할 수 있습니다.

### 1. `ngrok` 설치 및 인증

먼저 `ngrok`을 설치하고, `ngrok` 웹사이트에서 발급받은 인증 토큰을 설정해야 합니다.

````python
%%bash
# ngrok 다운로드 및 설치
curl -s https://ngrok-agent.s3.amazonaws.com/ngrok.asc | sudo tee /etc/apt/trusted.gpg.d/ngrok.asc >/dev/null
echo "deb https://ngrok-agent.s3.amazonaws.com/ apt main" | sudo tee /etc/apt/sources.list.d/ngrok.list
sudo apt update
sudo apt install -y ngrok
````

#### 1-1. `ngrok` 인증 토큰 설정

1.  `ngrok` 웹사이트 (https://ngrok.com/) 에 접속하여 계정을 생성하거나 로그인합니다.
2.  대시보드에서 'Your Authtoken' 섹션으로 이동하여 `ngrok` 인증 토큰을 복사합니다.
3.  아래 코드 셀의 `YOUR_NGROK_AUTHTOKEN` 부분을 복사한 토큰으로 대체하여 실행합니다.

````python
%%bash
# 여기에 발급받은 ngrok 인증 토큰을 입력하세요.
NGROK_AUTHTOKEN="YOUR_NGROK_AUTHTOKEN"
ngrok authtoken $NGROK_AUTHTOKEN
````

### 2. `ngrok` 터널 실행

이제 `ngrok`을 사용하여 Apache 웹 서버가 실행 중인 80번 포트에 대한 터널을 엽니다. 이 셀을 실행하면 `ngrok`이 외부 접속 가능한 URL을 제공합니다.

````python
%%bash --bg --output ngrok_output.txt
# Apache가 기본으로 사용하는 80번 포트에 대해 ngrok 터널을 엽니다.
ngrok http 80
````

### 3. `ngrok` URL 확인 및 접속

`ngrok` 터널이 시작된 후 몇 초 뒤에 생성된 URL을 확인할 수 있습니다. 이 URL을 웹 브라우저에 입력하여 `phpinfo.php` 페이지에 접속해 보세요.

````python
import time
import re

time.sleep(5) # ngrok이 완전히 시작될 시간을 기다립니다.

with open('ngrok_output.txt', 'r') as f:
    output = f.read()

# ngrok 출력에서 Public URL을 추출합니다.
match = re.search(r'"url":"(https?://[a-zA-Z0-9.-]+\.ngrok-free\.app)"', output)

if match:
    public_url = match.group(1)
    phpinfo_url = f"{public_url}/phpinfo.php"
    print(f"ngrok Public URL: {public_url}")
    print(f"PHP Info URL: {phpinfo_url}")
    print("이 URL을 웹 브라우저에 붙여넣어 Apache와 PHP 연동을 확인하세요.")
else:
    print("ngrok Public URL을 찾을 수 없습니다. ngrok이 제대로 시작되지 않았거나 인증 토큰이 잘못되었을 수 있습니다.")
    print("ngrok 출력:")
    print(output)
````

## NFS (Network File System) 서버 및 클라이언트 설정

NFS는 네트워크를 통해 파일 시스템을 공유하고 접근할 수 있도록 해주는 분산 파일 시스템 프로토콜입니다. 다른 컴퓨터에 있는 파일을 마치 로컬 파일처럼 사용할 수 있게 해줍니다. 여기서는 NFS 서버를 설정하고, 클라이언트에서 해당 서버에 마운트하는 방법을 설명합니다.

### 1. NFS 서버 설정

#### 1-1. NFS 서버 패키지 설치

먼저 NFS 서버 역할을 할 시스템에 `nfs-server` 패키지를 설치합니다.

````python
%%bash
sudo apt update
sudo apt install -y nfs-server
````

#### 1-2. 공유할 디렉토리 생성 및 설정

클라이언트에게 공유할 디렉토리를 생성하고 적절한 권한을 설정합니다. 여기서는 `/srv/nfs_share` 디렉토리를 생성합니다.

````python
%%bash
sudo mkdir -p /srv/nfs_share
sudo chown nobody:nogroup /srv/nfs_share
sudo chmod 777 /srv/nfs_share
echo "/srv/nfs_share 디렉토리가 생성되고 권한이 설정되었습니다."
````

#### 1-3. `/etc/exports` 파일 설정

`/etc/exports` 파일은 어떤 디렉토리를 어떤 클라이언트에게 어떤 권한으로 공유할지 정의합니다. 다음 예시에서는 `127.0.0.1` (로컬호스트)에게 `/srv/nfs_share` 디렉토리를 읽고 쓸 수 있는 권한(`rw`)과 추가 옵션(`sync`, `no_subtree_check`, `no_root_squash`)으로 공유합니다.

**주의:** Colab 환경에서 `127.0.0.1`은 현재 VM 자신을 의미합니다. 다른 VM이나 외부에서 접근하려면 해당 VM의 실제 IP 주소를 사용해야 하지만, Colab VM의 외부 접근은 일반적으로 제한됩니다.

````python
%%bash
sudo bash -c 'echo "/srv/nfs_share 127.0.0.1(rw,sync,no_subtree_check,no_root_squash)" >> /etc/exports'
cat /etc/exports
echo "/etc/exports 파일이 업데이트되었습니다."
````

#### 1-4. NFS 서버 재시작 및 내보내기 설정 적용

설정을 적용하기 위해 NFS 커널 서버를 재시작하고, `exportfs -a` 명령어로 변경 사항을 즉시 반영합니다.

````python
%%bash
sudo systemctl restart nfs-kernel-server
sudo exportfs -a
sudo systemctl status nfs-kernel-server
echo "NFS 서버가 재시작되고 설정이 적용되었습니다."
````

#### 1-5. 공유된 파일 시스템 확인

`showmount -e` 명령어를 사용하여 서버에서 어떤 디렉토리를 공유하고 있는지 확인할 수 있습니다.

````python
%%bash
sudo showmount -e 127.0.0.1
````

### 2. NFS 클라이언트 설정

#### 2-1. NFS 클라이언트 패키지 설치

NFS 서버에 접속할 클라이언트 시스템에 `nfs-common` 패키지를 설치합니다.

````python
%%bash
sudo apt update
sudo apt install -y nfs-common
````

#### 2-2. 마운트 포인트 생성

NFS 서버의 공유 디렉토리를 마운트할 로컬 디렉토리(마운트 포인트)를 생성합니다. 여기서는 `/mnt/nfs_client` 디렉토리를 사용합니다.

````python
%%bash
sudo mkdir -p /mnt/nfs_client
echo "/mnt/nfs_client 디렉토리가 생성되었습니다."
````

#### 2-3. NFS 공유 마운트

`mount` 명령어를 사용하여 NFS 서버의 공유 디렉토리를 로컬 마운트 포인트에 마운트합니다. `[NFS_SERVER_IP]`는 NFS 서버의 IP 주소입니다. Colab 환경에서는 `127.0.0.1`을 사용하여 동일 VM 내에서 테스트합니다.

````python
%%bash
sudo mount 127.0.0.1:/srv/nfs_share /mnt/nfs_client
echo "NFS 공유가 마운트되었습니다."
````

#### 2-4. 마운트 확인

`df -h` 명령어로 마운트된 파일 시스템 목록을 확인하거나, `mount` 명령어로 특정 마운트 포인트를 확인할 수 있습니다. 공유 디렉토리에 파일을 생성하여 쓰기 권한을 테스트해볼 수도 있습니다.

````python
%%bash
df -h | grep nfs_client

# 테스트 파일 생성
echo "Hello from NFS client!" | sudo tee /mnt/nfs_client/test_file.txt
cat /srv/nfs_share/test_file.txt

echo "NFS 마운트 및 쓰기 테스트가 완료되었습니다."
````

#### 2-5. 부팅 시 자동 마운트 설정 (옵션)

시스템 재부팅 후에도 NFS 공유가 자동으로 마운트되도록 하려면 `/etc/fstab` 파일에 내용을 추가해야 합니다. 이 작업은 `Colab` 환경에서는 임시적이므로, VM이 재시작되면 사라집니다. 실제 서버 환경에서 사용합니다.

````python
%%bash
sudo bash -c 'echo "127.0.0.1:/srv/nfs_share /mnt/nfs_client nfs defaults 0 0" >> /etc/fstab'
cat /etc/fstab
echo "/etc/fstab 파일이 업데이트되었습니다. 실제 운영 환경에서는 IP 주소를 NFS 서버의 실제 IP로 변경해야 합니다."
````

#### 2-6. 마운트 해제 (cleanup)

NFS 공유를 더 이상 사용하지 않거나, 설정을 변경하기 전에 마운트를 해제하려면 `umount` 명령어를 사용합니다.

````python
%%bash
sudo umount /mnt/nfs_client
df -h | grep nfs_client
echo "NFS 공유가 해제되었습니다."
````

## Samba를 이용한 Windows-Linux 간 파일 공유 (Linux 서버, Windows 클라이언트)

Samba는 Linux/Unix 시스템에서 Windows 운영 체제와 파일 및 프린터 공유를 할 수 있도록 해주는 소프트웨어입니다. 이를 통해 Linux 서버의 특정 디렉토리를 Windows 클라이언트에서 네트워크 드라이브처럼 사용할 수 있습니다.

### 1. Samba 서버 (Linux) 설정

먼저 리눅스 서버에서 Samba를 설치하고 공유 설정을 합니다.

#### 1-1. Samba 패키지 설치

Linux 서버에 `samba` 패키지를 설치합니다.

````python
%%bash
sudo apt update
sudo apt install -y samba
또는
sudo apt install samba samba-common-bin
````

#### 1-2. 공유할 디렉토리 생성 및 권한 설정

Windows 클라이언트와 공유할 디렉토리를 생성하고, Samba 사용자가 접근할 수 있도록 적절한 권한을 부여합니다. 여기서는 `/samba_share` 디렉토리를 생성하겠습니다.

````python
%%bash
sudo mkdir -p /samba_share
sudo chmod -R 0777 /samba_share
sudo chown nobody:nogroup /samba_share
echo "/samba_share 디렉토리가 생성되고 권한이 설정되었습니다."
````

#### 1-3. Samba 설정 파일 편집 (`/etc/samba/smb.conf`)

Samba의 메인 설정 파일인 `/etc/samba/smb.conf`를 편집하여 공유할 디렉토리를 정의합니다. 기존 파일을 백업하고 새 설정을 추가하는 것이 좋습니다.

여기서는 간단한 공유 설정을 추가할 것입니다. `[global]` 섹션 아래에 다음 내용을 추가하거나, 파일 끝에 새 섹션 `[shared_folder]`를 추가합니다.

**주의:** Colab 환경에서는 VM이 외부에서 직접 접근할 수 있는 공용 IP를 제공하지 않으므로, 일반적인 시나리오를 설명하지만 실제 외부 Windows에서 접속하려면 네트워크 환경 설정이 필요합니다.

````python
%%bash
# 기존 smb.conf 파일 백업
sudo cp /etc/samba/smb.conf /etc/samba/smb.conf.backup

# 새로운 공유 설정 추가
sudo bash -c 'cat << EOF >> /etc/samba/smb.conf

[shared_folder]
   comment = Samba Shared Folder
   path = /samba_share
   browseable = yes
   writable = yes
   guest ok = no
   read only = no
   create mask = 0777
   directory mask = 0777
   force user = nobody
   force group = nogroup
EOF'

echo "Samba 설정 파일이 업데이트되었습니다."
cat /etc/samba/smb.conf
````

#### 1-4. Samba 사용자 계정 생성

Samba는 시스템 사용자 계정을 사용하지만, 별도의 Samba 비밀번호를 설정해야 합니다. 기존 Linux 사용자가 없다면, `useradd`로 사용자를 생성하고 `smbpasswd`로 Samba 비밀번호를 설정합니다. 여기서는 `sambauser`라는 계정을 만들고 비밀번호를 설정해 보겠습니다.

**주의:** `smbpasswd` 명령은 대화형으로 비밀번호를 입력받습니다. Colab에서는 직접 입력해야 합니다.

````python
%%bash
# 시스템 사용자 생성 (sambauser가 없는 경우)
sudo useradd -M -s /sbin/nologin sambauser || true # 이미 존재하면 무시

# Samba 비밀번호 설정 (대화형으로 비밀번호 입력 필요)
echo -e "your_samba_password\nyour_samba_password" | sudo smbpasswd -a sambauser

echo "Samba 사용자 'sambauser'가 생성되었습니다. (비밀번호는 'your_samba_password'로 설정됨)"
````

#### 1-5. Samba 서비스 재시작

설정 변경 사항을 적용하려면 Samba 서비스를 재시작해야 합니다.

````python
%%bash
sudo systemctl restart smbd nmbd
sudo systemctl status smbd nmbd
echo "Samba 서비스가 재시작되었습니다."
````

### 2. Windows 클라이언트에서 Samba 공유 접근

이제 Samba 서버 설정이 완료되었으므로, Windows 클라이언트에서 공유 폴더에 접근하는 방법을 설명합니다.

**중요:** Colab 환경에서는 VM이 외부에서 직접 접근할 수 있는 IP 주소를 제공하지 않으므로, 아래 설명은 일반적인 리눅스 서버 환경에서 Windows 클라이언트가 접속하는 방법입니다. Colab 내에서는 이 단계를 직접 테스트하기 어렵습니다.

1.  **네트워크 드라이브 매핑:**
    *   Windows 탐색기를 열고 왼쪽 메뉴에서 '네트워크'를 클릭합니다.
    *   상단 메뉴에서 '네트워크 드라이브 연결'을 선택합니다.
    *   드라이브 문자(예: `Z:`)를 선택하고, 폴더 경로에 `\\<Linux 서버 IP 주소>\shared_folder` (예: `\\192.168.1.100\shared_folder`)를 입력합니다.
    *   '다른 자격 증명으로 연결'을 선택하고, 리눅스 서버에서 설정한 Samba 사용자(`sambauser`)와 비밀번호(`your_samba_password`)를 입력합니다.

2.  **직접 경로 접근:**
    *   Windows 탐색기 주소창에 `\\<Linux 서버 IP 주소>\shared_folder`를 직접 입력하고 Enter를 누릅니다.
    *   사용자 이름과 비밀번호를 묻는 창이 나타나면, Samba 사용자 계정 정보를 입력합니다.

**참고:**
*   `<Linux 서버 IP 주소>`는 실제 리눅스 서버의 내부 네트워크 IP 주소여야 합니다. Colab VM의 IP는 `ifconfig` 명령으로 확인할 수 있지만, 이는 Colab 내부 네트워크에서만 유효합니다.
*   Windows 방화벽이나 리눅스 서버의 방화벽(ufw, iptables)이 Samba 포트(TCP 445, 139)를 차단하고 있지 않은지 확인해야 합니다.

이러한 단계를 통해 Windows 클라이언트에서 Linux 서버의 공유 폴더에 접근하고 파일을 읽고 쓸 수 있게 됩니다.

### 1. Windows 서버 설정

Windows PC에 공유 폴더를 만들고 네트워크에서 접근할 수 있도록 설정해야 합니다. Colab 환경에서는 Windows OS를 직접 제어할 수 없으므로, 다음 단계를 수동으로 진행해야 합니다.

        *   '암호로 보호된 공유 끄기' (테스트 시 편리, 보안상 비권장) 또는 공유에 사용할 Windows 사용자 계정의 암호를 기억합니다.

### 2. Linux 클라이언트 설정

Windows 공유 폴더를 마운트할 로컬 디렉토리를 생성합니다. 여기서는 `/mnt/windows_share`를 사용합니다.

````python
%%bash
sudo mkdir -p /mnt/windows_share
echo "/mnt/windows_share 디렉토리가 생성되었습니다."
````

#### 2-3. Windows 공유 폴더 마운트

`mount` 명령어를 사용하여 Windows 공유 폴더를 마운트합니다. `[WINDOWS_SERVER_IP]`는 Windows PC의 IP 주소로 대체하고, `[WINDOWS_SHARE_NAME]`은 Windows에서 공유한 폴더 이름(예: `SharedFolder`)으로 대체합니다. `username`과 `password`는 Windows 공유에 접근할 때 사용되는 Windows 사용자 계정 정보입니다.

**주의:** Colab 환경에서는 외부 Windows PC에 직접 접근할 수 없습니다. 이 코드는 Colab VM과 동일한 네트워크에 있는 Windows PC가 있을 때만 작동합니다. 따라서 이 단계는 실제 로컬 네트워크 환경에서 테스트해야 합니다.

````python
%%bash
# 예시: sudo mount -t cifs //[WINDOWS_SERVER_IP]/[WINDOWS_SHARE_NAME] /mnt/windows_share -o username=your_windows_username,password=your_windows_password,uid=$(id -u),gid=$(id -g)
# 실제 Windows IP와 공유 이름, 사용자 정보로 대체하세요.

# 예시: 로컬 네트워크에 192.168.1.100 이라는 Windows 서버가 있고 'SharedFolder'를 공유했으며, 사용자 이름이 'Guest'이고 비밀번호가 없는 경우 (보안상 비권장)
# sudo mount -t cifs //192.168.1.100/SharedFolder /mnt/windows_share -o guest,uid=$(id -u),gid=$(id -g)

echo "Windows 공유 마운트 명령어를 실행할 준비가 되었습니다. 위의 주석 처리된 예시를 참고하여 실제 환경에 맞게 수정하세요."
````

#### 2-4. 마운트 확인

`df -h` 명령어로 마운트된 파일 시스템 목록을 확인하거나, 마운트 포인트(`ls /mnt/windows_share`)를 확인하여 Windows 공유 폴더의 내용이 보이는지 확인합니다.

#### 2-5. 부팅 시 자동 마운트 설정 (옵션)

시스템 재부팅 후에도 Windows 공유가 자동으로 마운트되도록 하려면 `/etc/fstab` 파일에 다음 내용을 추가할 수 있습니다. 이 작업은 `Colab` 환경에서는 임시적이므로, VM이 재시작되면 사라집니다. 실제 서버 환경에서 사용합니다.

**주의:** `fstab` 설정 시 잘못된 옵션은 시스템 부팅 문제를 일으킬 수 있으므로 주의해야 합니다. 특히, `_netdev` 옵션을 추가하여 네트워크가 활성화된 후에 마운트되도록 하는 것이 좋습니다.
